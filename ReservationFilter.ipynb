```python
import pandas as pd
import random

# File paths
# input_file = r"D:\Jupyter_projects\Salesforce_dataset\master_data_filtering\ReservationFilter.xlsx"
# output_file = r"D:\Jupyter_projects\Salesforce_dataset\master_data_filtering\ReservationFilter_Masked.xlsx"

D:\Jupyter_projects\Salesforce_dataset\master_data_25112024
input_file = r"D:\Jupyter_projects\Salesforce_dataset\master_data_25112024\ReservationFilter.xlsx"
output_file = r"D:\Jupyter_projects\Salesforce_dataset\master_data_25112024\ReservationFilter_Masked.xlsx"


# Pools for generating dummy addresses and last names
street_names = [
    "Main St", "Broadway", "Elm St", "Oak Ave", "Maple Dr", "Pine St", 
    "Cedar Blvd", "Birch Rd", "Spruce Ct", "Hickory Ln"
]

last_names = [
    "Smith", "Johnson", "Brown", "Taylor", "Anderson", "Lee", "Clark", "Lewis", 
    "Martinez", "Garcia", "Rodriguez", "Walker", "Hall", "Allen", "Young", 
    "King", "Wright", "Lopez", "Hill", "Scott", "Green", "Adams", "Baker"
]

# Masking functions
def generate_unit_address(value):
    """Generate a readable dummy address."""
    if pd.isnull(value):
        return value
    house_number = random.randint(1, 9999)
    street = random.choice(street_names)
    return f"{house_number} {street}"

def generate_guest_last_name(value):
    """Generate a random last name for the guest."""
    if pd.isnull(value):
        return value
    return random.choice(last_names)

# Load the Excel file
try:
    df = pd.read_excel(input_file)

    # Apply masking for Unit_Address__c
    if "Unit_Address__c" in df.columns:
        df["Unit_Address__c"] = df["Unit_Address__c"].apply(generate_unit_address)
    else:
        print("Warning: Column 'Unit_Address__c' not found in the dataset.")

    # Apply masking for Guest_Last_Name_Welcome_Letter__c
    if "Guest_Last_Name_Welcome_Letter__c" in df.columns:
        df["Guest_Last_Name_Welcome_Letter__c"] = df["Guest_Last_Name_Welcome_Letter__c"].apply(generate_guest_last_name)
    else:
        print("Warning: Column 'Guest_Last_Name_Welcome_Letter__c' not found in the dataset.")

    # Save the modified file
    df.to_excel(output_file, index=False, engine='openpyxl')  # Save as Excel file
    print(f"Masked file saved to: {output_file}")
except Exception as e:
    print(f"Error processing file: {e}")
```
